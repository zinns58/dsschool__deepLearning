## Define Sigmoid

```python
def sigmoid(n):
    return 1 / (1 + np.exp(-n))
```

```python
xx = np.linspace(-10, 10)
yy = sigmoid(xx)

plt.plot(xx, yy)
```

## Single layer Neural Network

```python
def gradient_descent(X, y, num_epoch=100, learning_rate=1.0):
    num_features = X.shape[1]
    
    w = np.random.uniform(low=-1.0, high=1.0, size=num_features)
    b = np.random.uniform(low=-1.0, high=1.0)
    
    for epoch in range(num_epoch):
        y_predict = X.dot(w) + b
        y_predict = sigmoid(y_predict)
        
        predict = y_predict >= 0.5
        accuracy = (predict == y).mean()
        
        print("{} accuracy = {:.5f}".format(epoch, accuracy))
        
        if accuracy == 1.00:
            break
            
        w = w - learning_rate * (y_predict - y).dot(X)
        b = b - learning_rate * (y_predict - y).mean()
        
    return w, b, accuracy
```

## And Data

```python
data = and_data.copy()

X = data[["x1", "x2"]].values
y = data["y"].values

w, b, accuracy = gradient_descent(X, y)

print("---" * 10)
print("accuracy = {0:.6f}".format(accuracy))
```

```python
w1, w2 = w[0], w[1]

xx = np.linspace(start=0.0, stop=1.0)
yy = -1.0 * (w1 * xx + b) / w2

plt.scatter(data['x1'], data['x2'], c=data['y'])
plt.plot(xx, yy, c='r')
```

## Multi-layer Neural Network

```python
data = xor_data.copy()

X = data[["x1", "x2"]].values
y = data["y"].values

y = y.reshape(4, 1)

y[0:10]
```

```python
num_epoch = 300
learning_rate = 1.0

w1 = np.random.uniform(low=-1.0, high=1.0, size=(2, 3))
b1 = np.random.uniform(low=-1.0, high=1.0, size=(1, 3))

w2 = np.random.uniform(low=-1.0, high=1.0, size=(3, 1))
b2 = np.random.uniform(low=-1.0, high=1.0, size=(1, 1))

for epoch in range(num_epoch):

    # Forward Propagation
    z1 = X.dot(w1) + b1
    a1 = sigmoid(z1)
    z2 = a1.dot(w2) + b2
    a2 = sigmoid(z2)
    
    y_predict = a2
    
    predict = y_predict > 0.5
    accuracy = (predict == y).mean()
    
    if accuracy == 1.0:
        break
        
    if epoch % 10 == 0:
        print("{:3} accuracy = {:.6f}".format(epoch, accuracy))
        
    # Backward Propagation
    d2 = a2 - y
    d1 = d2.dot(w2.T) * a1 * (1 - a1)
    
    w2 = w2 - learning_rate * a1.T.dot(d2)
    w1 = w1 - learning_rate * X.T.dot(d1)
    
    b2 = b2 - learning_rate * d2.mean(axis=0)
    b1 = b1 - learning_rate * d1.mean(axis=0)
    
print("---" * 10)
print("{:3} accuracy = {:.6f}".format(epoch, accuracy))
```
