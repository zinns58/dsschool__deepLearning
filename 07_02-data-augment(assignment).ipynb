```python
import numpy as np
import pandas as pd
```

## Load Dataset

```python
train = pd.read_csv("../08-07/data/train.csv")

print(train.shape)
train.head()
```

## Load Image

```python
from scipy.ndimage import imread

def load_images(filenames):
    images = []
    
    for filename in filenames:
        image = imread(filename)
        images.append(image)
        
    images = np.array(images)
    
    return images
```

```python
filenames = "../08-07/data/train/" + train["filename"]

X_train = load_images(filenames)
X_train.shape
```

```python
y_train = train["target"].values

print(y_train.shape)
y_train[0:10]
```

## Visualize

```python
%matplotlib inline
import matplotlib.pyplot as plt

figure, axes = plt.subplots(nrows=2, ncols=5)
figure.set_size_inches(24, 8)

print(y_train[0:10])

axes[0][0].imshow(X_train[0])
axes[0][1].imshow(X_train[1])
axes[0][2].imshow(X_train[2])
axes[0][3].imshow(X_train[3])
axes[0][4].imshow(X_train[4])
axes[1][0].imshow(X_train[5])
axes[1][1].imshow(X_train[6])
axes[1][2].imshow(X_train[7])
axes[1][3].imshow(X_train[8])
axes[1][4].imshow(X_train[9])
```

## Preprocessing

```python
from keras.preprocessing.image import ImageDataGenerator

generator = ImageDataGenerator(rotation_range=45,
                              horizontal_flip=True,
                              width_shift_range=0.3,
                              height_shift_range=0.3,
                              zoom_range=(0.7, 1.3),
                              vertical_flip=True)
```

```python
original_image = X_train[0]

transform_images = [generator.random_transform(original_image) for _ in range(10)]

figure, axes = plt.subplots(nrows=2, ncols=5)
figure.set_size_inches(24, 8)

axes[0][0].imshow(transform_images[0])
axes[0][1].imshow(transform_images[1])
axes[0][2].imshow(transform_images[2])
axes[0][3].imshow(transform_images[3])
axes[0][4].imshow(transform_images[4])
axes[1][0].imshow(transform_images[5])
axes[1][1].imshow(transform_images[6])
axes[1][2].imshow(transform_images[7])
axes[1][3].imshow(transform_images[8])
axes[1][4].imshow(transform_images[9])
```
